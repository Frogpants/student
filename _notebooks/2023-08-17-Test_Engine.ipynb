```python
import turtle as t
import math as m

def glob_var():
    global camx
    global camy
    global camz
    global rotx
    global roty
    global detail
    global t_index
    global t_width
    global t_height
    global resolution
    global points
    global texture
    detail = False
    set_texture()
    points = []
    t_index = 1
    t_width = 16
    t_height = 16
    resolution = 11
    camx = 0
    camy = 0
    camz = -200
    rotx = 0
    roty = 0

def set_texture():
    texture = []
    if detail == True:
        t = open("_notebooks/RESULTS (1).txt","r")
        t = t.read(256)
        t = t.splitlines(True)
        texture = t
    else:
        t = (0,0,0)
        while len(texture) < 256:
            texture.append(t)

def point(x,y,z):
    global px
    global py
    global pz

    px = x
    py = y
    pz = z

def goto(x,y,z):
    point(x-camx,y-camy,z-camz)
    point((px*m.cos(rotx))-(pz*m.sin(rotx)),py, (px*m.sin(rotx))+(pz*m.cos(rotx)))
    point(px, point((py*m.cos(roty))-(pz*m.sin(roty)), (py*m.sin(roty))+(pz*m.cos(roty))))
    if (pz < 0.01):
        pz = 0.01
    point(250*(px/pz),250*(py/pz),pz)
    points.append(x)
    points.append(y)
    points.append(z)
    t.Turtle.goto(px,py)

def fill(x1,z1,u1,v1,x2,z2,u2,v2):
    global tmp

    if round(_x1,1) == round(_x2,1):
        return
    if abs(_x1) > 240 and abs(_x2):
        if (_x1 * _x2) > 0:
            return
    if abs(_x1) > 240:
        tmp = ((_x1 > 0)*480)-240
        _z1 = _z1 + (((_z2-_z1)/(_x1-_x2))*(_x1-tmp))
        _u1 = _u1 + (((_u2-_u1)/(_x1-_x2))*(_x1-tmp))
        _v1 = _v1 + (((_v2-_v1)/(_x1-_x2))*(_x1-tmp))
        _x1 = tmp
    if abs(_x2) > 240:
        tmp = ((_x2 > 0)*480)-240
        _z2 = _z1 + (((_z2-_z1)/(_x1-_x2))*(_x1-tmp))
        _u2 = _u1 + (((_u2-_u1)/(_x1-_x2))*(_x1-tmp))
        _v2 = _v1 + (((_v2-_v1)/(_x1-_x2))*(_x1-tmp))
        _x2 = tmp
    if (_u1 / _z1) < (_u2 / _z2):
        integer = m.ceil(_u1 / _z1)
        mul = 1
        add = 0
    else:
        integer = m.floor(_u1 / _z1)
        mul = -1
        add = 1
    
    if (_v1 / _z1) < (_v2 / _z2):
        integer2 = m.ceil(_v1 / _z1)
        mul2 = 1
        add2 = 0
    else:
        integer2 = m.floor(_v1 / _z1)
        mul2 = -1
        add2 = 1
    t.Turtle.goto(_x1,cury)
    t.pencolor(texture[t_index + (((integer - (1-add)) % t_width) + (((integer2-(1-add2)) % t_height) * t_wdith))])
    t.pendown()
    out3 = (_u1 - (_z1 * integer)) / (((_z2 - _z1) * _integer) - (_u2 - _u1))
    out2 = (_v1 - (_z1 * integer2)) / (((_z2 - _z1) * _integer2) - (_v2 - _v1))
    lim1 = (_u2/_z2) * mul
    lim2 = (_v2/_z2) * mul2
    if integer * mul > lim1:
        out3 = "infinity"
    if integer2 * mul2 > lim2:
        out2 = "infinity"
    if out3 != "NaN":
        while out3 == "infinity" and out2 == "infinity":
            if out2 > out3:
                t.Turtle.xcor(_x1+((_x2-_x1)*out3))
                t.pencolor(texture[(((integer - add) % t_width) + (((integer2 - mul2 - add2) % t_height) * t_width))])
                integer += mul
                out3 = (_u1 - (_z1 * integer)) / (((_z2 - _z1) * integer) - (_u2 - _u1))
                if integer * mul > lim1:
                    out3 = "infinity"
            else:
                t.Turtle.xcor(_x1+((_x2-_x1)*out2))
                t.pencolor(texture[(((integer - add) % t_width) + (((integer2 - mul2 - add2) % t_height) * t_width))])
                integer2 += mul2
                out2 = (_u1 - (_z1 * integer2)) / (((_z2 - _z1) * integer2) - (_u2 - _u1))
                if integer2 * mul2 > lim2:
                    out2 = "infinity"
        t.Turtle.xcor(_x2)
        t.penup()
        _x1 = x1
        _z1 = z1
        _u1 = u1
        _v1 = v1
        _x2 = x2
        _z2 = z2
        _u2 = u2
        _v2 = v2

def rfill(iterations):
    global _x1
    global _x2
    global _z1
    global _z2
    global _u1
    global _u2
    global _v1
    global _v2
    global cury

    while cury < iterations:
        fill(_x1,_z1,_u1,_v1,_x2,_z2,_u2,_v2)
        _x1 += vx1
        _x2 += vx2
        _z1 += vz1
        _z2 += vz2
        _u1 += vu1
        _u2 += vu2
        _v1 += vv1
        _v2 += vv2
        cury += resolution

def text_tri(x_1,y_1,z_1,u_1,v_1,x_2,y_2,z_2,u_2,v_2,x_3,y_3,z_3,u_3,v_3):
    global vx1
    global vx2
    global vz1
    global vz2
    global vu1
    global vu2
    global vv1
    global vv2

    ylimit = m.ceil(-180/resolution) * resolution
    tmp2 = []
    tmp2.append(x_1)
    tmp2.append(x_2)
    tmp2.append(x_3)
    tmp2.append(y_1)
    tmp2.append(y_2)
    tmp2.append(y_3)
    tmp2.append(1/z_1)
    tmp2.append(1/z_2)
    tmp2.append(1/z_3)
    tmp2.append(u_1/z_1)
    tmp2.append(u_2/z_2)
    tmp2.append(u_3/z_3)
    tmp2.append(v_1/z_1)
    tmp2.append(v_2/z_2)
    tmp2.append(v_3/z_3)
    t.pensize(resolution)
    minC = 4
    if tmp2[5] < tmp2[minC]:
        minC = 5
    if tmp2[6] < tmp2[minC]:
        minC = 6
    maxC = 4
    if tmp2[5] > tmp2[maxC]:
        maxC = 5
    if tmp2[6] > tmp2[maxC]:
        maxC = 6
    midC = 4
    while not (midC == maxC or midC == minC):
        midC += 1
    _x1 = tmp2[minC-3]
    _x2 = tmp2[minC-3]
    _z1 = tmp2[minC+3]
    _z2 = tmp2[minC+3]
    _u1 = tmp2[minC+6]
    _u2 = tmp2[minC+6]
    _v1 = tmp2[minC+9]
    _v2 = tmp2[minC+9]
    cury = m.ceil(tmp2[minC]/resolution)*resolution
    if cury < ylimit:
        cury = ylimit
    tmp = resolution / (cury - tmp2[midC])
    tmp_2 = (cury - tmp2[minC]) / (tmp2[midC] - tmp2[minC])
    _x1 = tmp2[minC-3] + ((tmp2[midC-3] - tmp2[minC]) * tmp_2)
    vx1 = (_x1 - tmp2[midC-3]) * tmp
    _z1 = tmp2[minC+3] + ((tmp2[midC+3] - tmp2[minC+3]) * tmp_2)
    vz1 = (_z1 - tmp2[midC+3]) * tmp
    _u1 = tmp2[minC+6] + ((tmp2[midC+6] - tmp2[minC+6]) * tmp_2)
    vu1 = (_u1 - tmp2[midC+6]) * tmp
    _v1 = tmp2[minC+9] + ((tmp2[midC+9] - tmp2[minC+9]) * tmp_2)
    vv1 = (_v1 - tmp2[midC+9]) * tmp

    tmp = resolution / (cury - tmp2[maxC])
    tmp_2 = (cury - tmp2[minC]) / (tmp2[maxC] - tmp2[minC])
    _x2 = tmp2[minC-3] + ((tmp2[maxC-3] - tmp2[minC]) * tmp_2)
    vx2 = (_x1 - tmp2[maxC-3]) * tmp
    _z2 = tmp2[minC+3] + ((tmp2[midC+3] - tmp2[minC+3]) * tmp_2)
    vz2 = (_z1 - tmp2[maxC+3]) * tmp
    _u2 = tmp2[minC+6] + ((tmp2[midC+6] - tmp2[minC+6]) * tmp_2)
    vu2 = (_u1 - tmp2[maxC+6]) * tmp
    _v2 = tmp2[minC+9] + ((tmp2[midC+9] - tmp2[minC+9]) * tmp_2)
    vv2 = (_v1 - tmp2[maxC+9]) * tmp
    if tmp2[midC] > 180:
        rfill(180)
    else:
        rfill(tmp2[midC])
    tmp = resolution / (cury - tmp2[maxC])
    tmp_2 = (cury - tmp2[midC]) / (tmp2[maxC] - tmp2[midC])
    _x1 = tmp2[midC-3] + ((tmp2[maxC-3] - tmp2[midC]) * tmp_2)
    vx1 = (_x1 - tmp2[maxC-3]) * tmp
    _z1 = tmp2[midC+3] + ((tmp2[maxC+3] - tmp2[midC+3]) * tmp_2)
    vz1 = (_z1 - tmp2[midC+3]) * tmp
    _u1 = tmp2[midC+6] + ((tmp2[maxC+6] - tmp2[midC+6]) * tmp_2)
    vu1 = (_u1 - tmp2[maxC+6]) * tmp
    _v1 = tmp2[midC+9] + ((tmp2[maxC+9] - tmp2[midC+9]) * tmp_2)
    vv1 = (_v1 - tmp2[maxC+9]) * tmp
    if tmp2[maxC] > 180:
        rfill(180)
    else:
        rfill(tmp2[maxC])

def quad(x1,y1,z1,x2,y2,z2,x3,y3,z3,x4,y4,z4):
    text_tri(x1,y1,z1,0,16,x2,y2,z2,16,16,x3,y3,z3,0,0)
    text_tri(x2,y2,z2,16,16,x3,y3,z3,0,0,x4,y4,z4,16,0)

def quad_3D(x1,y1,z1,x2,y2,z2,x3,y3,z3,x4,y4,z4):
    points = []
    goto(x1,y1,z1)
    goto(x2,y2,z2)
    goto(x3,y3,z3)
    goto(x4,y4,z4)
    quad_3D(points[0],points[1],points[2],points[3],points[4],points[5],points[6],points[7],points[8],points[9],points[10],points[11],points[12])

# Put all Objects in the render function to be drawn every frame

def render():
    quad_3D(-50,-50,100,50,-50,100,-50,50,100,50,50,100)

# Prepare all vars
glob_var()

# Create Scene

while True:
    t.update()

    render()

    t.exitonclick()
```
